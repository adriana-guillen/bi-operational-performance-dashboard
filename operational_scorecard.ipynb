## 2. Business Context.

Gas station operations were evaluated through a structured performance framework composed of three primary sources:


a) **ADOPTA una Estación**: internal weekly evaluations conducted by multidisciplinary staff, assessing service quality, cleanliness, execution, and operational standards.  
b) **Operative Evaluation**: technical inspections focused on infrastructure, safety, and compliance of operational processes.  
c) **Mystery Shopper**: third-party evaluations capturing the customer experience during real service interactions.


Each station received a consolidated performance score based on these evaluations, forming the basis for monthly performance reviews conducted at both station and commercial assessor levels.

This evaluation system directly influenced operational decisions and employee performance, including variable compensation linked to service quality outcomes.

Despite the structured nature of the evaluations, performance data remained distributed across sources, making it difficult to obtain a unified and comparable view of service quality across regions, stations, and responsible roles.

As a result, **identifying performance gaps**, tracking trends over time, and enforcing accountability **required manual consolidation and interpretation of data.**

### 2.1 Evaluation Framework Overview.

The operational performance system is built on three complementary evaluation sources, each capturing a different dimension of service quality.

| Evaluation Source   | Evaluator Type | Primary Focus Area |
|--------------------|----------------|--------------------|
| ADOPTA una Estación | Internal staff | Operational execution, cleanliness, service standards |
| Operative Evaluation | Technical team | Infrastructure, safety, and compliance |
| Mystery Shopper     | External third party | Customer experience during real service interactions |

Each source contributes to a consolidated performance score, forming a standardized basis for comparison across stations, regions, and commercial assessors.

## 3. Analytical Objective.

The objective of this project is to enable **consistent, comparable, and actionable monitoring of service performance** across gas station operations.

Specifically, the solution is designed to:

a) Standardize evaluation outputs into a single performance metric on a 0–100 scale.  
b) Consolidate multiple evaluation sources into a unified analytical view.  
c) Allow segmentation by **region, station, evaluation type, and commercial assessor.**  
d) Support time-based analysis to track performance trends and detect variability.  
e) Make performance **directly attributable to responsible roles** to strengthen accountability.  

The goal is to replace fragmented, source-level analysis with a **single decision framework** that supports faster identification of performance gaps and clearer operational follow-up.

## 4. Approach.

### 4.1 Data Structure.

The solution is built on a structured data model designed to represent operational evaluations at the station level.

The model separates transactional data from contextual attributes to support consistent aggregation and filtering across multiple operational dimensions.

### 4.1.1 Data Model Components.

a) **Fact Table (Evaluations):**  
  Contains individual evaluation records, including date, evaluation type, station identifier, and performance score.

b) **Dimension Table (Stations):**  
  Provides contextual attributes such as region, station name, and assigned commercial assessor.



### 4.1.2 Main Fields in the Evaluation Dataset.

| Column Name        | Description |
|--------------------|-------------|
| Date               | Week of evaluation |
| Evaluation Type    | Source of evaluation (ADOPTA, Operative, Mystery Shopper) |
| Gas Station Number | Unique identifier of each station |
| Grade              | Evaluation score (0–100 scale) |

This structure enables consistent comparison of service performance across time, regions, stations, and responsible roles.

### 4.2 KPI Design.

Service performance is standardized into a single metric:

a) **Score (0–100 scale)**: represents the consolidated result of operational evaluations.  
b) Aggregations are calculated as averages to support comparison across:
  - Time (monthly and yearly trends).  
  - Region.  
  - Commercial assessor.  
  - Evaluation type.  

This standardization ensures comparability across heterogeneous evaluation sources.


### 4.3 Dashboard Architecture.

The dashboard is designed as a decision interface structured into three analytical layers:

a) **Control Layer (Filters).**  
  Enables dynamic segmentation by region, station, commercial assessor, evaluation type, and time period.

b) **Trend Analysis (Time-Based Monitoring).**  
  Visualizes performance evolution over time to identify patterns, variability, and potential declines in service quality.

c) **Accountability Layer (Performance by Commercial Assessor).**  
  Displays performance at the responsible role level, enabling direct comparison and ownership tracking.

d) **Structural Comparison (Region and Evaluation Type).**  
  Allows identification of systemic differences across regions and evaluation sources.

e) **Executive Summary View (Yearly Trend).**  
  Provides a simplified, high-level view of performance evolution for decision-makers.
